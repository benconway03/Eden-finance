```python
import yfinance as yf
import pandas as pd
import IPython
from datetime import datetime, timedelta, timezone
```

```python
def curr_prices():
    tickers = ["AAPL", "MSFT", "GOOGL", "AMZN"]
    cryptos = ["BTC-GBP", "ETH-GBP", "SOL-GBP"]

    # --- Stocks ---
    ticker_data = yf.download(tickers, interval="1d", period="1d")
    ticker_close_prices = ticker_data["Close"]

    # Transpose: tickers as rows, date as column header
    ticker_transposed = ticker_close_prices.T

    # Keep only the last date column and reset index
    last_date = ticker_transposed.columns[-1]
    ticker_transposed = ticker_transposed[[last_date]].reset_index()
    ticker_transposed.columns = ["Ticker", "Price T"]

    # --- Cryptos ---
    now = datetime.now(timezone.utc)
    midnight = now.replace(hour=0, minute=0, second=0, microsecond=0)
    start = midnight
    end = midnight + timedelta(minutes=5)

    crypto_data = yf.download(cryptos, start=start, end=end, interval="1m")
    crypto_closes = crypto_data["Close"]

    # First non-missing value for each crypto
    first_crypto_prices = crypto_closes.bfill().iloc[0].reset_index()
    first_crypto_prices.columns = ["Ticker", "Price T"]

    # --- Stack stocks and crypto vertically ---
    all_prices = pd.concat([ticker_transposed, first_crypto_prices], ignore_index=True)

    # --- Display ---

    print("All prices combined:")
    display(all_prices)

curr_prices()
```
